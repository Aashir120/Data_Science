```python
import pandas as pd
```

```python
df = pd.read_csv('Mobile_Price_Classifiation_train_missing.csv')
```

```python
#i. Find which attribute(s) have null values
df.isnull().sum()
```

```python
#ii. Drop a column titled “Unnamed: 0” as it is useless 
df.drop('Unnamed: 0', axis=1,inplace=True)
df.isnull().sum()
```

```python
#iii. Display only those data not having NaN concerning battery_power 
nn = df.dropna(subset=['battery_power'])
print(nn['battery_power'])
```

```python
#iv. Drop all data corresponding to NaN using the “dropna” method with inplace parameter. 
df.dropna(inplace=True)
df.isnull().sum()
```

```python
#v. Read the file again, and drop all data corresponding to NaN concerning the ram. 
df = pd.read_csv('Mobile_Price_Classifiation_train_missing.csv')
nn = df.dropna(subset=['ram'])
print(nn['ram'])
```

```python
# vi. Read the file again, and fill the NaN with the mean or mode values of corresponding
# attributes with the inplace parameter. 
df = pd.read_csv('Mobile_Price_Classifiation_train_missing.csv')
df.fillna({'battery_power':df['battery_power'].mean(), 'ram':df['ram'].mode()}, inplace=True)
print(df['ram'])
print(df['battery_power'])
```

```python
#vii. Read the file again, and fill the NaN in the “battery_power” attribute group by the
#“n_cores” attribute. 
df = pd.read_csv('Mobile_Price_Classifiation_train_missing.csv')
df['battery_power'] = df.groupby(['n_cores'])['battery_power'].fillna(method='ffill')
print(df['battery_power'])
```

```python
#viii. Fill the NaN in the “ram” attribute group by the “price_range” attribute. 
df = pd.read_csv('Mobile_Price_Classifiation_train_missing.csv')
df['ram'] = df.groupby(['price_range'])['ram'].fillna(method='ffill')
print(df['ram'])
```

```python
#ix. Read the file again, and fill the NaN with a constant value. 
df = pd.read_csv('Mobile_Price_Classifiation_train_missing.csv')
df.fillna(0,inplace=True)
df
```

```python
#x. Suggest the best fill of missing value of attributes battery_power and ram according to the dataset 

#Ans: according to dataset, The best method to fill NaN values might be forward fill because of the variation in 
# computational power of batter power and ram. so ffill would be best in this case
df = pd.read_csv('Mobile_Price_Classifiation_train_missing.csv')
df['ram'].fillna(method='ffill' ,inplace=True)
df['battery_power'].fillna(method='ffill' ,inplace=True)
print(df['ram'])
print(df['battery_power'])
```
